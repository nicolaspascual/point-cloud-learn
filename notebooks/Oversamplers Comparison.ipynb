```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = (20,10)
plt.style.use('seaborn-whitegrid')
```

```python
OVERSAMPLERS = """
PLAIN SMOTE (SVC): 0.594573612661848 +/- 0.1268699147745886
ADASYN (SVC): 0.6029069459951812 +/- 0.13387503940239076
SPIDER (LinearDiscriminantAnalysis): 0.42363323415490595 +/- 0.09299719313002884
SWIM (SVC): 0.5393207282913165 +/- 0.08456512533231204
polynom_fit_SMOTE (SVC): 0.6007051282051282 +/- 0.1373442725818786
ProWSyn (SVC): 0.5783918695683401 +/- 0.1234872059964299
SMOTE_IPF (SVC): 0.6029069459951812 +/- 0.13387503940239076
Lee (SVC): 0.6029069459951812 +/- 0.13387503940239076
SMOBD (SVC): 0.6029069459951812 +/- 0.13387503940239076
G_SMOTE (SVC): 0.6128977884860237 +/- 0.12629442583563216
CCR (SVC): 0.5821507199370977 +/- 0.15077919597872955
LVQ_SMOTE (SVC): 0.5839939961998786 +/- 0.10897107842537938
Assembled_SMOTE (SVC): 0.594573612661848 +/- 0.1268699147745886
SMOTE_TomekLinks (SVC): 0.6029069459951812 +/- 0.13387503940239076
"""
```

```python
labels, vals, errs = [], [], []

for res in OVERSAMPLERS.strip().split('\n'):
    name, numbers = res.split(':')
    val, err = numbers.split('+/-')
    val, err = float(val), float(err)
    labels.append(name)
    vals.append(val)
    errs.append(err)
```

```python
x = list(range(len(labels)))
plt.bar(x, vals, yerr=errs, label='F1')
plt.xticks(x, labels, rotation=30)
plt.ylim(0.3, 0.8)
plt.legend()
```
